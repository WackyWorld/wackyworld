# 作業二 yahoo爬蟲

```python
from bs4 import BeautifulSoup
```

```python
# 使用 BeautifulSoup 
urls="https://tw.stock.yahoo.com/quote/2888.TW"
context = requests.get(urls)
soup = BeautifulSoup(context.text,'lxml')

# 抓取股票table資料
ul_tag = soup.find('ul', class_='D(f)')
```

```python
span_items = ul_tag.find_all('span')
y_fields = []
y_data = []
space = 0 #遇到漲跌符號的情況
for index, item in enumerate(span_items):
    if item.text == '':
        space += 1
        continue
    
    print(item.text)

    #去掉空格
    if space %2 == 1:
        print("space")
        if index % 2 == 1:
            y_fields.append(item.text)
        else:
            y_data.append(item.text)
        
    else:
        if index % 2 == 0:
            y_fields.append(item.text)
        else:
            y_data.append(item.text)

print("Fields:", y_fields)
print("Data:", y_data)
```

```python
yfinance = pd.DataFrame([y_data],columns=y_fields)
```
